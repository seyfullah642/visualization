```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df_url = 'https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/athlete_events.csv'
df = pd.read_csv(df_url)
```

```python
df
```

```python
df = df.loc[df['Year'] == 2016]
```

```python
df
```

```python
df.Sport.unique()
df.Medal.unique()
```

```python
df.Medal.count()
```

```python
df.dropna(0, subset=['Medal'], inplace=True)
```

```python
df
```

```python
df.Sport.value_counts()
```

```python
top_sports = ['Athletics', 'Swimming', 'Rowing', 'Football', 'Hockey']
```

```python
df = df[df.Sport.isin(top_sports)]
```

```python
sns.catplot(x='Sport', data=df, aspect=1.5, kind='count')
```

```python
sns.displot(df.Age, kde=False)
```

```python
plot = sns.catplot(x='Team', data=df, aspect=3, kind='count')
plot.set_xticklabels(rotation=90)
```
