```python
import seaborn as sns
```

```python
mpg_df = sns.load_dataset('mpg')
```

```python
mpg_df.head()
```

```python
ax = sns.scatterplot(x='weight', y='mpg', data=mpg_df)
```

```python
sns.set(style='ticks')
sns.jointplot(x='weight', y='mpg', data=mpg_df, kind='hex', color='#4CB391')
```

```python
sns.set_style('white')
sns.kdeplot(x='weight', y='mpg', data=mpg_df, shade=True)
```

```python
sns.set_style('white')
ax1 = sns.scatterplot(x='model_year', y='mpg', data=mpg_df)
```

```python
ax = sns.lineplot(x='model_year', y='mpg', data=mpg_df, ci=68)
```
